```python
## Data Preparation -- Building Permits from FRED 
import pandas as pd
import numpy as np
```

```python
building_permits = pd.read_csv('./data/building_permits.csv')
```

```python
reported_columns = ['Survey_Date', 'FIPS_State', 'FIPS_County', 'Region_Code',
       'Division_Code', 'State', 'state_abbrev', 'County_Name', 'one_unit_buildings_rep',
       'one_unit_units_rep', 'one_unit_value_reps', 'two_unit_buildings_rep',
       'two_unit_units_rep', 'two_unit_value_rep', 'three_unit_building_rep',
       'three_unit_untis_rep', 'three_unit_value_rep',
       'five_unit_building_rep', 'five_unit_unit_rep', 'five_unit_value_rep']
building_permits = building_permits[reported_columns]
```

```python
building_permits = building_permits[building_permits['Survey_Date'] == 202110]
```

```python
zip_code_data = pd.read_csv('https://nycdsacapstone2021.blob.core.windows.net/additionaldata/ZIP_all_data.csv', index_col=0)
zip_code_data = zip_code_data.reset_index()
```

```python
building_permits['County_Name'] =building_permits['County_Name'].str.replace("County", "").str.upper().str.strip()
```

```python
final = pd.merge(zip_code_data, building_permits, how = 'left', left_on = ['County', 'State'], right_on = ['County_Name', 'state_abbrev'])
```

```python
final = final.rename(columns = {'State_x' : 'State'})
```

```python
final.to_csv('final_zip_data.csv', index = False)
```
