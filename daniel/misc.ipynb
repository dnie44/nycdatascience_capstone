```python
hospitals = pd.read_csv('_data/Hospitals.csv', index_col='Facility ID')
hospitals.shape
```

```python
hosp_Addr = hospitals.Address+', '+hospitals.City+', USA'
```

```python
import geopy
from geopy import Nominatim
from geopy.extra.rate_limiter import RateLimiter

locator = Nominatim(user_agent="nycDSA")
geocode = RateLimiter(locator.geocode, min_delay_seconds=0.5)
location = locator.geocode("Champ de Mars, Paris, France")
location[1]
```

```python
locations = [geocode(addr) for addr in hosp_Addr]
```

```python
Addr_coords = [each[1] if each!=None else np.NaN for each in locations]
Addr_coords = pd.DataFrame(Addr_coords, index=hospitals.index, columns=['latitude', 'longitude'])
```

```python
hospitals = hospitals.join(Addr_coords)
```

for Missing: Use Hospital Name to find Lat-Long 

```python
hospitals['Facility Name'] = hospitals['Facility Name'].apply(lambda each: str.title(each))
```

```python
hosp_missing = hospitals[hospitals.latitude.isna()][['Facility Name','City']]
hosp_missing = hosp_missing['Facility Name']+', '+hosp_missing.City+', USA'
```

```python
locations = [geocode(h_name) for h_name in hosp_missing]
```

```python
Addr_coords = [each[1] if each!=None else np.NaN for each in locations]
```

```python
# Manually fill first coord for Marshall Medical Center, BOAZ, USA
Addr_coords[0] = geocode('Marshall Medical Center, BOAZ, USA')[1]
```

```python
Addr_coords = pd.DataFrame(Addr_coords, index=hosp_missing.index, columns=['latitude', 'longitude'])
```

```python
for each in Addr_coords.index:
    hospitals.loc[each,'latitude'] = Addr_coords.loc[each,'latitude']
    hospitals.loc[each,'longitude'] = Addr_coords.loc[each,'longitude']
```
