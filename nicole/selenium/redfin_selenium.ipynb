#### Instruct driver to download CSVs

```python
#grab csv files for those respective zipcodes

null_zipcodes = {}

for zipcode in zipcodes:
    
    try:

        driver.get(f'https://www.redfin.com/zipcode/{zipcode}/filter/include=forsale+mlsfsbo+construction+fsbo+sold-3yr')
        time.sleep(1)

        download = driver.find_element(By.ID,'download-and-save')
    
        download.click()
        time.sleep(1.5)
    
    except Exception as e:

        print(f"Invalid URL at {zipcode}")
        null_zipcodes[zipcode] = zipcode
```

## Cleaning & saving the scraped data

```python
import os
import glob

#set the desired working directory
os.chdir("../raw_data_redfin/")

class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
extension = 'csv'

#list of all file names in directory
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames])

file_name = '../combined_data/combined_redfin_' + datetime.now().strftime("%m-%d-%y") + '.csv'

#save to to csv
combined_csv.to_csv(file_name, index=False, encoding='utf-8-sig')
```

```python
#rename csv files to mark them as scanned 
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
[os.rename(f, f[:-4]+'_scanned.csv') for f in all_filenames]
```
